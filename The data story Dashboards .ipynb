```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('largest_companies.csv')

# Fixed: Changed 'Rank' to 'Ranks' to match your column name
if df.iloc[0]['Ranks'] == 'Ranks':
    df = df.drop(0)

# Make sure Revenue and Profit are numbers
# Note: Use the exact column names from your Task 2 (e.g., 'Revenue' and 'Profit')
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Quick check to see the columns
print(df.columns)
```

```python
#pie chrt representation
industry_counts = df['Industry'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Companies by Industry')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Revenue', y='Profit', data=df, hue='Industry', s=100)

plt.title('Business Efficiency: Revenue vs. Profit')
plt.xlabel('Revenue (Billions USD)')
plt.ylabel('Profit (Billions USD)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# Sort by revenue and take the top 10
top_10_revenue = df.sort_values(by='Revenue', ascending=False).head(10)

plt.figure(figsize=(12, 8))

# Create the bar chart
sns.barplot(x='Revenue', y='Name', data=top_10_revenue, palette='magma')

# Add professional labels and title
plt.title('Top 10 Global Companies by Revenue (in Billions USD)', fontsize=15)
plt.xlabel('Revenue (USD Billions)', fontsize=12)
plt.ylabel('Company Name', fontsize=12)

# Add a grid to make the numbers easier to read
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
